```python
# Merging two dataframes

merge_df = df.set_index('employee_id').join(df1.set_index('EMPLOYEE #'))
merge_df = merge_df.reset_index()
```

```python
# Filling the NaN values with mean values for the respective column

merge_df.describe()
merge_df.fillna(merge_df.mean(), inplace = True)
```

```python
merge_df.isnull().any()
```

```python
# Removing Employee number

merge_df = merge_df.drop(columns = 'employee_id')
```

```python
# Checking how many employees left

merge_df['left'].value_counts()
```

```python
# Checking Correlation

merge_df.corr()
```

```python
# Perform One Hot Encoding on Categorical Data

categorical = ['department', 'salary']
merge_df = pd.get_dummies(merge_df, columns = categorical)
```

```python
merge_df.shape
```

```python
# Preparing our Dataset for Machine Learning

from sklearn.model_selection import train_test_split

X = merge_df.drop(['left'], axis = 1).values
y = merge_df['left'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Normalize the Data

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
df_train = pd.DataFrame(X_train)
df_train.head(2)
```

```python
# Training a Logistic Regression Model

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

model = LogisticRegression(solver = 'lbfgs')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Accuracy {0:.2f}%".format(100*accuracy_score(y_pred, y_test)))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Training a Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

score = model.score(X_test, y_test)

print("Accuracy {0:.2f}%".format(100*accuracy_score(y_pred, y_test)))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Training on Deep Neural Networks

import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

model.add(Dense(9, kernel_initializer="uniform", activation="relu", input_dim=20))
model.add(Dense(1, kernel_initializer="uniform", activation="sigmoid"))


model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
model.summary()
```

```python
# Start training the Classifier

batch_size = 10
epochs = 25

history = model.fit(X_train, 
                    y_train, 
                    batch_size = batch_size, 
                    epochs = epochs, 
                    verbose = 1, 
                    validation_data = (X_test, y_test))

score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# Displaying the Classification Report and Confusion Matrix

y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
